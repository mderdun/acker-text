```python
import xml.etree.ElementTree as ET
import xml.dom.minidom

# Open the text file in read mode
with open('Sourcebook/completeSade.txt', 'r') as file:
    # Read the contents of the file
    contents = file.read()

# Split the contents into individual paragraphs
paragraphs = contents.split('\n')

# Filter out empty paragraphs
paragraphs = [p for p in paragraphs if p.strip()]

# Create an XML element tree structure
root = ET.Element('root')

# Initialize a variable to hold the current paragraph
current_paragraph = ""

# Iterate over the paragraphs and create a new XML element for each paragraph
for paragraph in paragraphs:
    # If the paragraph starts with a lowercase letter, append it to the current paragraph
    if paragraph and paragraph[0].islower():
        # If the current paragraph ends with a '-', remove it and merge the paragraphs without a space
        if current_paragraph and current_paragraph[-1] == '-':
            current_paragraph = current_paragraph[:-1] + paragraph
        else:
            current_paragraph += " " + paragraph
    else:
        # If the current paragraph is not empty, create a new XML element for it
        if current_paragraph:
            paragraph_element = ET.SubElement(root, 'p')
            paragraph_element.text = current_paragraph

        # Start a new current paragraph
        current_paragraph = paragraph

# If there is a remaining current paragraph after the loop, create a new XML element for it
if current_paragraph:
    paragraph_element = ET.SubElement(root, 'p')
    paragraph_element.text = current_paragraph

# Convert the ElementTree to a string
xml_string = ET.tostring(root, 'utf-8')

# Parse the XML string with minidom
dom = xml.dom.minidom.parseString(xml_string)

# Prettify the XML string
pretty_xml = dom.toprettyxml()

# Write the pretty XML string to a file
with open('xml/sources/completeSade.xml', 'w') as file:
    file.write(pretty_xml)
```
